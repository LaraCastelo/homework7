```python
import requests 
from bs4 import BeautifulSoup
response = requests.get("https://www.lemonde.fr/")
doc = BeautifulSoup(response.text,'html.parser')
```

```python
#We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.
```

```python
# make a list of all the artiles
```

```python
#dê um nome para a lista
```

```python
articles = doc.find_all('div', class_= 'article')
```

```python
#peça o primeiro item para ver se está certo.
```

```python
articles[0]
```

```python
len(articles)
```

```python
#faça um for loop para ir para todos os artigos dentro da classe "articles"
#vai precisar fazer uma lista > rows []
#fazer dicionários row {}
```

```python
rows = []
for art in articles:
     row = {}
     try:
        row['title'] = art.select_one('.article__title').get_text()
        
     except:
        print("Headline not found!")
     try:
        row["link"] = art.find('a')['href']
     except:
         print("Couldn't find a link")
     try:
        row["premium"] = art.find(class_='sr-only').get_text()
     except:
        print("It's free to read")
     try:
         row["article_type"] = art.find(class_='article__type').get_text()
     except:
         print("No article type")
     try:
         row['byline'] = art.find(class_='article__byline').get.text()
     except:
         print("Couldn't find a bylne!")
     try:
         row["image"]=art.find('img')['src']
         
     except:
        print("No image")
     
     print(row)
     rows.append(row)
```

```python
# transfor in a df > so you can use pandas
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(rows)
df
```

```python
#salve como csv
```

```python
df.to_csv("lemonde-headlines.csv")
```
